```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV

warnings.filterwarnings("ignore")
```

```python
data = load_iris()
```

```python
X = data.data
y = data.target
```

```python
X
```

```python
y
```

```python
# Splitting data into train, test:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
model = DecisionTreeClassifier(criterion="entropy") # By default criterion is gini.
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.tree import plot_tree
```

```python
plot_tree(model, filled = True)
```

```python
y_pred = model.predict(X_test)
```

```python
# Importing Evaluation Metrics:
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred)}")
print(f"ConfusionMatrix : {confusion_matrix(y_test, y_pred)}")
print(f"ClassificationReport : \n{classification_report(y_test, y_pred)}")
```

```python
# Post pruning:
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(criterion = "gini", max_depth = 3)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred)}")
print(f"ConfusionMatrix : {confusion_matrix(y_test, y_pred)}")
print(f"ClassificationReport : \n{classification_report(y_test, y_pred)}")
```

```python
# Hyper Parameter Tuning(Prepruning):
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {"criterion" : ["gini", "entropy", "log_loss"],
          "splitter" : ["best", "random"],
          "max_depth" : [1, 2, 3, 4, 5, 6, 7, 10],
          "max_features" : ["sqrt", "log2", "auto"]
         }
```

```python
grid = GridSearchCV(model, param_grid = params, cv = 5, scoring = "accuracy", verbose = 3)
```

```python
grid.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
grid.best_params_ # This will give us the best parameters.
```

```python
print(f"Accuracy : {accuracy_score(y_test, y_pred)}")
print(f"ConfusionMatrix : {confusion_matrix(y_test, y_pred)}")
print(f"ClassificationReport : \n{classification_report(y_test, y_pred)}")
```
